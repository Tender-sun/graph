### 生成一个简单的图数据（无权值）


```python
import matplotlib
import networkx as nx
from PIL import Image
import matplotlib.pyplot as plt
from networkx.drawing.nx_pydot import to_pydot
from matplotlib.font_manager import *
g=nx.Graph()#创建空的无向图
```

下面是图的构建过程，规定了15个节点和自定义的一些边。下面是无权重的建设过程

```python
nodes = ['a','b','c','d','e','f','g','h','i','g','k','l','m','n','o']
edges = [('a','b'),('a','c'),('a','g'),('a','e'),('b','d'),('b','e'),('b','m'),('d','n'),('e','i'),('m','n'),('g','h'),('h','i'),('a','f'),('c','f'),('f','o'),('f','l'),('l','k'),('k','d')]
g.add_nodes_from(nodes)#添加节点
#g.add_weighted_edges_from(edges)#添加边,这种方式是附带权重的结果
g.add_edges_from(edges) #这种方式添加的边是没有权值的边数据
```

```python
nx.draw(g,
        with_labels=True,
        edge_color = 'r',
        node_color = 'g',
        node_size = 300,
)
```

这次生成一些带权值的

```python
g2 = nx.Graph() #创建一个新的无向图
nodes2 = ['a','b','c','d','e','f','g','h','i','g','k','l','m','n','o']
edges2 = [('a','b',1),('a','c',2),('a','g',2),('a','e',8),('b','d',4),('b','e',6),('b','m',4),('d','n',7),('e','i',1),('m','n',3),('g','h',1),('h','i',5),('a','f',3),('c','f',1),('f','o',2),('f','l',3),('l','k',1),('k','d',4)]
g2.add_nodes_from(nodes2)#添加节点
g2.add_weighted_edges_from(edges2)#添加边,这种方式是附带权重的结果
#g2.add_edges_from(edges2) #这种方式添加的边是没有权值的边数据
```

```python
edges22,weights = zip(*nx.get_edge_attributes(g2,'weight').items())
```

```python

nx.draw(g2,
        with_labels=True,
        node_color = 'g',
        node_size = 300,
        edge_color = weights,
        width = 3.0,
        edge_cmap=plt.cm.Blues,
)
```
